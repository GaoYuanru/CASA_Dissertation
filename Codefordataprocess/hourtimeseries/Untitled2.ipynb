```python
import pandas as pd

# 假设你有一个数据框 df，其中包含两列 'var1' 和 'var2'
df = pd.read_csv('merged_counts.csv')

# 计算所有变量之间的相关性
correlation_matrix = df.corr()

# 打印相关性矩阵
print(correlation_matrix)


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# 创建一个热图
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["axes.grid"] = False

f = plt.figure(figsize=(10, 10))

plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=90)
plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)

corr = df.corr().values
for i in range(len(corr)):
    for j in range(len(corr)):
        text = plt.text(j, i, np.round(corr[i, j], 2),
                       ha="center", va="center", color="black")

plt.show()
```

```python
import pandas as pd

# 读取csv文件
df = pd.read_csv('merged_counts.csv')

# 先转换为日期时间格式，然后再格式化为所需格式
formatted_time = pd.to_datetime(df.loc[:137, 'time'], format='%Y-%m-%d %H:%M:%S').dt.strftime('%Y/%d/%m %H:%M:%S')

# 将格式化后的时间赋值给 df
df.loc[:137, 'time'] = formatted_time

# 保存修改后的DataFrame
df.to_csv('cc_modified.csv', index=False)

```
